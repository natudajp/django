```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# サンプルデータの作成 (実際のデータに合わせて修正してください)
data = {'Date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05',
                               '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10',
                               '2023-01-11', '2023-01-12', '2023-01-13', '2023-01-14', '2023-01-15',
                               '2023-01-16', '2023-01-17', '2023-01-18', '2023-01-19', '2023-01-20',
                               '2023-01-21', '2023-01-22', '2023-01-23', '2023-01-24', '2023-01-25',
                               '2023-01-26', '2023-01-27', '2023-01-28', '2023-01-29', '2023-01-30']),
        'Close': [100, 102, 101, 103, 105, 104, 106, 107, 108, 110,
                  109, 111, 112, 113, 115, 114, 116, 117, 118, 120,
                  119, 121, 122, 123, 125, 124, 126, 127, 128, 130]}
df = pd.DataFrame(data)
df.set_index('Date', inplace=True)

print("サンプルデータ:")
display(df.head())
```

```python
# テクニカル指標の計算 (例: 5日移動平均線と25日移動平均線)
df['5MA'] = df['Close'].rolling(window=5).mean()
df['25MA'] = df['Close'].rolling(window=25).mean()

print("\nテクニカル指標を追加したデータ:")
display(df.head())
```

```python
# テクニカル指標の表示
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['5MA'], label='5-Day Moving Average')
plt.plot(df['25MA'], label='25-Day Moving Average')
plt.title('Close Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
